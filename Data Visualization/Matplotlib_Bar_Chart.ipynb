# Bar Chart

**✅ What is a Bar Chart?**

A bar chart displays categorical data with rectangular bars.

The length/height of the bar represents the value or frequency of that category.

**Types of Bar Charts**
1. Vertical Bar Charts
2. Horizontal Bar Charts
3. Stacked Bar Charts
4. Grouped Bar Charts

**Advanrages of Bar Charts**
- Easy to Understand
- Effective Comparison
- Versatile
- Customizable

**Applications of Bar Charts**
- Business Analytics
- Financial Analysis
- Survey Results
- Project Management

**Example: Sales of Different Products**

🎯 Objective:

Visualize the sales of different products using a bar chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: Product names and their sales
products = ['Laptop', 'Tablet', 'Smartphone', 'Headphones', 'Smartwatch']
sales = [150, 100, 300, 120, 80]
```

```python
# Create a bar chart
plt.bar(products, sales, color='skyblue', edgecolor='black')

# Add title and axis labels
plt.title('Sales of Electronic Products (in Units)')
plt.xlabel('Product')
plt.ylabel('Units Sold')

# Add grid and display the plot
plt.grid(axis='y')
plt.show()
```

**Explanation of the Code**
- 1️⃣ products: List of product categories (x-axis).
- 2️⃣ sales: Corresponding sales data (y-axis).
- 3️⃣ plt.bar(...):

    - Creates vertical bars for each product.

    - color='skyblue': Sets bar color.

    - edgecolor='black': Adds a black border to the bars.

- 4️⃣ plt.title(), plt.xlabel(), plt.ylabel(): Add labels and title.
- 5️⃣ plt.grid(axis='y'): Horizontal grid lines for better readability.
- 6️⃣ plt.show(): Displays the chart.

**📊 Output Description**

You’ll see a bar chart where:

Each bar represents a product.

The height shows how many units were sold.

## Creating Bar Chart with Matplotlib

### 1. Creating a Bar Chart using `plt` Approach

```python
import matplotlib.pyplot as plt

# Sample data : categories vs values
categories = ['Category A', 'Category B', 'Category C']
values = [23,45,56]

# Creating a bar chart
plt.figure(figsize=(8,5))
plt.bar(categories, values, color='skyblue')

# Adding title and labels
plt.title('Sales by Category')
plt.xlabel('Categories')
plt.ylabel('Sales')

# Display the chart
plt.show()
```

### 2. Creating a Bar Chart using `ax` Approach

```python
import matplotlib.pyplot as plt

# Sample data : categories and their values
categories = ['Category A', 'Category B', 'Category C']
values = [23, 45, 56]

# Creating a figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# Creating a bar chart
ax.bar(categories, values, color='lightgreen')

# Adding title and labels
ax.set_title('Sales by Category')
ax.set_xlabel('Categories')
ax.set_ylabel('Sales')

# Display the chart
plt.show()
```

**Example: Students Enrolled in Courses**

🎯 Objective:

Visualize student enrollment in different courses using a bar chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: Courses and student enrollments
courses = ['Python', 'Data Science', 'Web Dev', 'AI/ML', 'Cyber Security']
students = [120, 95, 75, 110, 60]
```

```python
# Create the bar chart
plt.bar(courses, students, color='green', edgecolor='black')

# Add title and labels
plt.title('Student Enrollment in Different Courses')
plt.xlabel('Courses')
plt.ylabel('Number of Students')

# Add grid and show the chart
plt.grid(axis='y')
plt.show()
```

**Explanation**
- 1️⃣ courses: List of course names (x-axis).
- 2️⃣ students: Number of students enrolled in each course (y-axis).
- 3️⃣ plt.bar(...): Draws vertical bars for each course.
- 4️⃣ plt.title(), plt.xlabel(), plt.ylabel(): Provide context to the chart.
- 5️⃣ plt.grid(axis='y'): Adds horizontal grid lines.
- 6️⃣ plt.show(): Renders the final chart.

**📊 Output**

A bar chart where each bar shows how many students are enrolled in each course — making it easy to compare course popularity.
