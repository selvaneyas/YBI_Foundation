# Introduction to Matplotlib

Matplotlib is a cornerstone library in Python for creating static, interactive , animated vizualizations. Understanding and utilizing Matplotlib is Crucial for anyone looking to transform raw data into insightful, visual narratives

## What is Matplotlib?
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in python.

## Why use Matplotlib?
Matplotlib is favored for its ability to produce publication-quality plots with just a few lines of code.

##  Types of Charts and Their Syntax

### 1. Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 2. Bar Chart

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)

plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### 3. Histogram

```python
import numpy as np

data = np.random.randn(1000)

#print(data)

plt.hist(data,bins=30)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### 5. Pie Chart

```python
sizes = [15, 30, 45, 10]
labels = ['A','B','C','D']

plt.pie(sizes, labels=labels , autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### Choosing the Right Chart
- Line Plot : Use for trends over time or Continuous data
- Bar Chart : Ideal for compraring different categories or groups
- Histogram : Best for showing the distribution of a dataset
- Scatter Plot : Use to identify relationships or correlations between variables
- Pie Chart : Effective for showing proportions of a whole

## Two Ways to Plot in Matplotlib : `plt` and `ax` Approach

### A. `plt` Approach

 Straight forward and suitable for simple plots. uses the state based interface.

```python
plt.plot(x,y)
plt.title('Simple Plot')
plt.show
```

### B. `ax` Approach

object-oriented approach , provide more control and customization. recommended for complex visualizations.

```python
fig,ax = plt.subplots()

ax.plot(x,y)
ax.set_title('Object-Oriented Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```
