# Line Charts

**✅ What is a Line Chart?**

A line chart shows trends over time or ordered data points by connecting data points with a continuous line.

It’s ideal for time-series data, growth patterns, and comparisons.

**Key Components of a Line Chart**
1. Axes
2. Data Points
3. Lines
4. Labels

**Advantages of Line Charts**
1. Trend Visualization
2. Clarity
3. Comparative Analysis
4. Data Density

**Disadvantages of Line Charts**
1. Overlapping Lines
2. Data Interpretation
3. Complexity

**Applications of Line Charts**
1. Business
2. Economics
3. Science
4. Technology

**✅ Example: Monthly Sales Trend**

🎯 Objective:

Visualize monthly sales (in ₹1000s) over a year using a line chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: Months vs Sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [25, 28, 30, 35, 40, 38, 45, 50, 48, 42, 38, 33]
```

```python
# Create the line chart
plt.plot(months, sales, color='blue', marker='o', linestyle='-', linewidth=2)
```

```python
# Add title and axis labels
plt.title('Monthly Sales Trend (in ₹1000s)')
plt.xlabel('Month')
plt.ylabel('Sales (₹1000s)')

# Add grid and display the plot
plt.grid(True)
plt.show()
```

**Explanation of the Code**
- 1️⃣ months: List of month names (x-axis).
- 2️⃣ sales: Corresponding sales data (y-axis).
- 3️⃣ plt.plot(...):

    - color='blue': Line color

    - marker='o': Circle markers at each data point

    - linestyle='-': Solid line

    - linewidth=2: Line thickness

- 4️⃣ plt.title(), plt.xlabel(), plt.ylabel(): Add a title and axis labels.
- 5️⃣ plt.grid(True): Adds grid lines for better readability.
- 6️⃣ plt.show(): Displays the chart.

**📊 Output Description**

The line chart will show how sales increase, peak, and drop across the months.

Useful for identifying trends, seasonality, or business performance over time.



## Creating Line Charts With Matplotlib

### A. Using the `plt` Approach

```python
import matplotlib.pyplot as plt

# Sample data  : time vs values
time = [1,2,3,4,5]
values = [10,20,15,25,30]

# Create a line chart
plt.plot(time,values,marker='o')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Simple Line Chart')

# Display the chart
plt.show()
```

### B. Using the `ax` Approach

```python
import matplotlib.pyplot as plt

# Sample data : time vs values
time = [1, 2, 3, 4, 5]
values = [10, 20, 15, 25, 30]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot data
ax.plot(time, values, marker='o')

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Values')
ax.set_title('Line Chart with Ax')

# Display the chart
plt.show()
```

**Example: Sales Comparison of Product A vs Product B**

🎯 Objective:

Visualize and compare monthly sales trends of two products using a line chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [30, 35, 40, 42, 45, 50]
sales_B = [25, 28, 35, 37, 40, 48]
```

```python
# Create line chart for Product A
plt.plot(months, sales_A, label='Product A', color='blue', marker='o', linestyle='-')

# Create line chart for Product B
plt.plot(months, sales_B, label='Product B', color='green', marker='s', linestyle='--')

# Add title and labels
plt.title('Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales (₹1000s)')

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

**Explanation of the Code**
- 1️⃣ months: X-axis labels for months.
- 2️⃣ sales_A and sales_B: Y-axis data for Product A and

B sales.

- 3️⃣ plt.plot(...):


Product A:

- color='blue', marker='o', linestyle='-'

Product B:

- color='green', marker='s', linestyle='--'

- 4️⃣ plt.legend(): Adds a legend to differentiate the two products.
- 5️⃣ plt.grid(True): Adds grid lines for better readability.
- 6️⃣ plt.show(): Displays the chart.

**📊 Output Description**

You will see two lines:

Product A in blue with circle markers

Product B in green with square markers

The chart allows you to compare growth trends side-by-side.
