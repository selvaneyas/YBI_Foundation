## 1. Basics of K-means Clustering

K-means is an unsupervised learning algorithm used to partition data into k distinct clusters based on feature similarity. üß©

## 2. Theory Behind K-means Clustering

### 2.1 Initialization
- Choose k initial centroids randomly. üéØ

### 2.2 Assignment
- Assign each data point to the nearest centroid. üìç

### 2.3 Update
- Recalculate centroids as the mean of assigned points. üîÑ

### 2.4 Report
- Repeat assignment and update steps until convergence. üìà

## 3. Example Scenario

Imagine grouping customers based on purchasing behavior to target marketing campaigns effectively. üéØ

## 4. Implementation Example
```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Create KMeans instance
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.title('K-means Clustering')
plt.show()
```

````python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
````

````python
# Sample data : Annual spending on different categories(in thousamds)
data = {
    'Customer':['Aarav','Ishita','Rohan','Sneha','Neha','Vikram','Ananya','Arjun'],
    'Spending_on_Electronics':[15,20,14,25,30,18,22,24],
    'Spending_on_Clothing':[35,40,30,45,50,38,42,44]
}
df = pd.DataFrame(data)
df
````

````python
# Extracting features for clustering
X = df[['Spending_on_Electronics','Spending_on_Clothing']]
X
````

````python
# Applying K-means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
````

````python
# Adding cluster labels to the dataframe
df['Cluster'] = kmeans.labels_

print(df)
````

````python
# Visualizing the clusters
plt.scatter(df['Spending_on_Electronics'], df['Spending_on_Clothing'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Spending on Electronics (in thousands)')
plt.ylabel('Spending on Clothing (in thousands)')
plt.title('Customer Segmentation using K-means Clustering')
plt.colorbar(label='Cluster')
plt.show()
````
