# 📉 Basics of Linear Regression Model

Welcome to this comprehensive guide on Linear Regression! Let's explore the fundamentals, equations, types, and practical implementations with Python. 🚀

## 1. Introduction

Linear Regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. 📊

## 2. Equation of a Straight Line

The general equation of a straight line is:

$$ y = mx + c $$

Where:
- $y$: Dependent variable (response) 🎯
- $x$: Independent variable (predictor) 📈
- $m$: Slope of the line (rate of change) ↗️
- $c$: Intercept (value of y when x=0) 📍

## 3. Linear Regression Equation

The linear regression model can be expressed as:

$$ y = b_0 + b_1 x_1 + b_2 x_2 + \dots + b_n x_n + \epsilon $$

Where:
- $y$: Dependent variable (target) 🎯
- $b_0$: Intercept (constant term) 📍
- $b_1, b_2, \dots, b_n$: Coefficients for independent variables 📊
- $x_1, x_2, \dots, x_n$: Independent variables (features) 🧩
- $\epsilon$: Error term (residuals) ⚠️

### Dependent and Independent Variables
- **Dependent Variable**: The outcome or target variable we want to predict. 🎯
- **Independent Variables**: The input features or predictors used to explain the dependent variable. 🧩

## 4. Types of Linear Regression

### 4.1 Simple Linear Regression
- Involves one independent variable and one dependent variable. 📈
- Equation: $$ y = b_0 + b_1 x_1 + \epsilon $$

### 4.2 Multiple Linear Regression (Two Independent Variables)
- Involves two independent variables. 🧩🧩
- Equation: $$ y = b_0 + b_1 x_1 + b_2 x_2 + \epsilon $$

### 4.3 Multiple Linear Regression (More than Two Independent Variables)
- Involves more than two independent variables. 🧩🧩🧩
- Equation: $$ y = b_0 + b_1 x_1 + b_2 x_2 + \dots + b_n x_n + \epsilon $$

## 5. Least Squares Method

The least squares method estimates the coefficients $b_0, b_1, \dots, b_n$ by minimizing the sum of the squared differences between the observed and predicted values. 🔍

Mathematically:

$$ \min_{b_0, b_1, \dots, b_n} \sum_{i=1}^m (y_i - \hat{y}_i)^2 $$

Where $y_i$ is the actual value and $\hat{y}_i$ is the predicted value.

## 6. Limitations of Linear Regression
- Assumes a linear relationship between dependent and independent variables. 📉
- Sensitive to outliers. ⚠️
- Multicollinearity among independent variables can affect estimates. 🔄
- Cannot capture complex relationships. 🔍

## 7. Advantages of Linear Regression
- Simple to understand and interpret. 🧠
- Computationally efficient. ⚡
- Works well for linearly separable data. 📈
- Provides insight into relationships between variables. 🔍

## 8. Disadvantages of Linear Regression
- Poor performance with non-linear data. 📉
- Sensitive to outliers. ⚠️
- Assumes homoscedasticity (constant variance of errors). 📊
- Limited flexibility. 🔒

## 9. Applications of Linear Regression
- Predicting house prices 🏠
- Stock market forecasting 📈
- Risk assessment in finance 💰
- Medical research and epidemiology 🏥
- Marketing analytics 📊

## 10. Simple Linear Regression

### Formula
$$ y = b_0 + b_1 x + \epsilon $$

### Assumptions
- Linearity between dependent and independent variable 📈
- Independence of errors ⚖️
- Homoscedasticity (constant variance of errors) 📊
- Normality of error terms 🔄

### Advantages

### Disadvantages

### Sample Python Code
```python

```

### Explanation
- We create a simple linear regression model using scikit-learn.
- Fit the model on sample data.
- Predict and plot the regression line.

````python
# Importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

````python
# Generating Sample Data
np.random.seed(2529)
x = np.random.rand(100, 1) * 10  # Feature: Random values between 0 and 10
y = 4 + 3 * x + np.random.randn(100, 1) * 2  # Target: Linear relation with noise
````

````python
# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2529)
````

````python
# Creating a linear regression model
model = LinearRegression()
````

````python
# Fitting the model
model.fit(x_train, y_train)
````

````python
# Making predictions
y_pred = model.predict(x_test)
````

````python
# Visualizing the results
plt.scatter(x_test, y_test, color='blue', label='Actual Data')
plt.scatter(x_test, y_pred, color='red', label='Predicted Data')
plt.legend()

plt.title('Simple Linear Regression Predictions vs Actual Data')
plt.xlabel('Feature (x) : Independent Variable')
plt.ylabel('Target (y) : Dependent Variable')
plt.show()
````

### Example Code

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
````

````python
# Sample Dataset
data = {
    'X1' : [1,2,3,4,5],
    'X2' : [5,3,6,9,8],
    'X3' : [7,4,3,9,6],
    'Y' : [14,24,18,34,17]
}
df = pd.DataFrame(data)
df
````

````python
# Independent Variables
X = df[['X1', 'X2', 'X3']]
X
````

````python
# Dependent Variable
y = df['Y']
y
````

````python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2529)
````

````python
# Creating a linear regression model
model = LinearRegression()
````

````python
# Fitting the Linear Regression model
model.fit(X_train, y_train)
````

````python
# Making predictions
y_pred = model.predict(X_test)
````

````python
# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
````

````python
# Print detailed model performance analysis
print("Model Performance Analysis:")
print("-" * 50)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mse):.2f}")
print(f"R-squared Score: {r2}")
print("\nModel Coefficients:")
for i, coef in enumerate(model.coef_):
    print(f"Feature X{i+1}: {coef:.4f}")
print(f"Intercept: {model.intercept_:.4f}")

print("\nFeedback:")
print("- The high MSE (571.64) indicates significant prediction errors")
print("- The R-squared value being NaN suggests potential issues with the model")
print("- Possible improvements:")
print("  * Increase training data size")
print("  * Feature scaling/normalization")
print("  * Check for multicollinearity")
print("  * Consider non-linear relationships")
````
